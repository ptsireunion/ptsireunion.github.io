Ces relations sont approximables par méthode d'Euler :
$$E_x= - \frac{V[i,j+1]-V[i,j]}{a}$$

$$E_y = - \frac{V[i+1,j]-V[i,j]}{a}$$

où $a=1\mu m$ est la distance entre deux cases adjacentes.
