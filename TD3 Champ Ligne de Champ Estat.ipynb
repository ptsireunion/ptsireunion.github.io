### Bonus: Utilisation de la fonction `numpy.linalg.norm` à la place de la racine carrée

La fonction `numpy.linalg.norm` est utilisée pour calculer la norme d'un vecteur ou d'une matrice. La norme est une mesure de la taille ou de la longueur d'un vecteur, et elle peut être calculée de différentes manières selon le paramètre spécifié. Voici quelques utilisations courantes :

**Norme Euclidienne (ou norme L2)** : C'est la norme par défaut pour les vecteurs. Elle est calculée comme la racine carrée de la somme des carrés des éléments du vecteur. Pour un vecteur $\overrightarrow{v} = [v_1, v_2, \ldots, v_n] \$, la norme L2 est donnée par :

   $
   \|v\|_2 = \sqrt{v_1^2 + v_2^2 + \cdots + v_n^2}
   \$

   Exemple :
   ```python
   import numpy as np
   v = np.array([3, 4])
   norm = np.linalg.norm(v)  # Calcule la norme L2 par défaut
   print(norm)  # Affiche 5.0
